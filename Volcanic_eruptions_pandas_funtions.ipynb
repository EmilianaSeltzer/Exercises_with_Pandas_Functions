```python
#we have some values with errors that ar null, we will delete them because we can't do much with them.
```

```python
pd.Timestamp('today').normalize()
```

```python
df['Day_today']=pd.Timestamp('today').normalize()
```

```python
df['Day_today'].info()
```

```python
df['Day_today'] = df.Day_today.dt.date
df.Day_today.info()
```

```python
df['Last_Eruption']= df.Last_Eruption.dt.date
df.Last_Eruption.info()
```

```python
df.Day_today - df.Last_Eruption
```

```python
df['Days_from_eruption']= df.Day_today - df.Last_Eruption
```

```python
df['Days_from_eruption']= pd.to_timedelta(df['Days_from_eruption'], errors = 'coerce').dt.days #we want to see just the days, and remove the time
```
