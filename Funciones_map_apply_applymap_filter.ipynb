```python
import pandas as pd
```

```python
#uso el diccionario
data = {
    'Nombre': ['Juan', 'María', 'Carlos', 'Laura', '_', 'Luis', 'Ana', 'Roberto', 'Elena', '_'],
    'Edad': [35, 28, 42, 30, 45, 39, 29, 47, 33, 41],
    'Salario': [50000, 60000, 75000, 45000, 80000, 55000, 62000, 72000, 49000, 68000],
    'Departamento': ['Ventas', 'TI', 'Ventas', 'Recursos Humanos', 'TI', 'Recursos Humanos', 'TI', 'Ventas', 'Recursos Humanos', 'TI']
}
```

```python
#creo el df

df = pd.DataFrame(data)
```

```python
df
```

```python
#filtrar valores de salarios entre 49.000 y 55.000

salarios_filtrados = filter(lambda x: x>= 49000 and x <= 55000, df['Salario'])
```

```python
#queremos reemplazar los - en la columna de Nombres, usamos MAP sobre Nombre
df['Nombre']=df.Nombre.map( lambda x: 'N/A' if x =='_' else x)
df
```

```python
#pasar todo el df a STR, primero preguntamos cuales ya son str:
df.applymap(lambda x: isinstance(x,str))
```
