```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
sales = pd.read_csv('Francia_Alemania_procesados.csv')
sales.sample(10)
```

```python
sales.info ()
```

```python
sales.Returns.unique()
```

```python
sales[['Quantity','UnitPrice','UnitPriceXQuantity']].describe()
```

Unit price, quantity, and total value by row

```python
sns.displot(sales['UnitPrice'], bins=100, height=4, aspect=3)
sns.rugplot(sales['UnitPrice'], color = '#4c72b0', height =.05)
plt.show()
```

```python
sns.displot(sales['Quantity'], bins=75, height=4, aspect=2)
sns.rugplot(sales['Quantity'], color = '#4c72b0', height =.05)
plt.show()
```

```python
sns.displot(sales['UnitPriceXQuantity'], bins=75, height=4, aspect=2)
```

### Difference between sales and returns

```python
sns.pairplot(data = sales[sales.Returns =='Compras'],x_vars=['UnitPrice','Quantity','UnitPriceXQuantity'],y_vars= ['UnitPrice','Quantity','UnitPriceXQuantity'],hue ='Country')
```

```python
sns.pairplot(data = sales[sales.Returns =='Devoluciones'],x_vars=['UnitPrice','Quantity','UnitPriceXQuantity'],y_vars= ['UnitPrice','Quantity','UnitPriceXQuantity'],hue ='Country')
```

Countries comparitions

```python
sns.countplot(x='Returns',data= sales)
plt.show()
```

```python
sns.countplot(x='Returns',data= sales, hue='Country')
plt.show()
```

```python
print(sales.groupby(['Country','Returns']).UnitPrice.mean())

sns.barplot(x='Returns',y='UnitPrice', data=sales, hue='Country')
plt.show()
```

```python
print(sales.groupby(['Country','Returns']).Quantity.mean())

sns.barplot(x='Returns',y='Quantity', data=sales, hue='Country')
plt.show()
```

```python
print(sales.groupby(['Country','Returns']).UnitPriceXQuantity.mean())

sns.barplot(x='Returns',y='UnitPriceXQuantity', data=sales, hue='Country')
plt.show()
```

Outliers

```python
sns.boxplot(x='Country',y="UnitPriceXQuantity", data=sales)
plt.show()
```

```python
sns.catplot(x='Returns',y="UnitPriceXQuantity", data=sales, hue="Country")
plt.show()
```
